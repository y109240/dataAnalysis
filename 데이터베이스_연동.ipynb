```python
import tkinter as tk
from tkinter import font
from tkinter import ttk
import pymysql
import time

# DB 정보
conn = pymysql.connect(host="127.0.0.1", user="root", password="0000", db="cash_db", charset='utf8')
curs = conn.cursor()
# curs.execute("drop table if exists cal")
curs.execute("create table if not exists cal(num int auto_increment primary key, datetime DATETIME, item varchar(10), money int, sl char(4))")
# curs.execute("insert into cal values(null, '2024-04-04-13', '녹차', 4000, '수입')")
# curs.execute("insert into cal values(null, '2024-01-01-10', '장난감', 2000, '지출')")
conn.commit()

def bt1_click():
    bt1_w = tk.Toplevel()
    bt1_w.title("입력")
    bt1_w.geometry("400x200")
    
    bt1_year_lb=tk.Label(bt1_w, text='년도')
    bt1_year_en=tk.Entry(bt1_w)
    bt1_year_en.insert(0, t.tm_year)
    bt1_year_lb.grid(row=0, column=0)
    bt1_year_en.grid(row=0, column=1)

    bt1_month_lb=tk.Label(bt1_w, text='월')
    bt1_month_en=tk.Entry(bt1_w)
    bt1_month_en.insert(0, t.tm_mon)    
    bt1_month_lb.grid(row=1, column=0)
    bt1_month_en.grid(row=1, column=1)

    bt1_day_lb=tk.Label(bt1_w, text='일')
    bt1_day_en=tk.Entry(bt1_w)
    bt1_day_en.insert(0, t.tm_mday)    
    bt1_day_lb.grid(row=2, column=0)
    bt1_day_en.grid(row=2, column=1)

    bt1_hour_lb=tk.Label(bt1_w, text='시간')
    bt1_hour_en=tk.Entry(bt1_w)
    bt1_hour_en.insert(0, t.tm_hour)
    bt1_hour_lb.grid(row=3, column=0)
    bt1_hour_en.grid(row=3, column=1)

    bt1_item_lb=tk.Label(bt1_w, text='항목')
    bt1_item_en=tk.Entry(bt1_w)
    bt1_item_lb.grid(row=4, column=0)
    bt1_item_en.grid(row=4, column=1)
 
    bt1_money_lb=tk.Label(bt1_w, text='금액')
    bt1_money_en=tk.Entry(bt1_w)
    bt1_money_lb.grid(row=5, column=0)
    bt1_money_en.grid(row=5, column=1)

    combobox = ttk.Combobox(bt1_w, values=["수입", "지출"])
    combobox.grid(row=6, column=0)
    
    def bt1_bt_click():
        bt1_en_year=bt1_year_en.get()
        bt1_en_month=bt1_month_en.get()
        bt1_en_day=bt1_day_en.get()
        bt1_en_hour=bt1_hour_en.get()
        bt1_en_item=bt1_item_en.get()
        bt1_en_money=bt1_money_en.get()
        selected_item = combobox.get()
        curs.execute("insert into cal values(null, '%s', '%s', %d, '%s')" % (bt1_en_year+'-'+bt1_en_month+'-'+bt1_en_day+'-'+bt1_en_hour,\
                                                                       bt1_en_item, int(bt1_en_money), selected_item))
        conn.commit()


    bt1_bt=tk.Button(bt1_w, text='저장', command=bt1_bt_click)
    bt1_bt.grid(row=5, column=2)
    

def bt2_click():
    bt2_w = tk.Toplevel()
    bt2_w.title("조회")
    bt2_w.geometry("400x200")

    lbox=tk.Listbox(bt2_w, width=40, height=10)
    lbox.pack()
    curs.execute("select * from cal order by num desc")
    rows = curs.fetchall()
    for row in rows:
         lbox.insert(0, row)



def bt3_click():
    bt3_w = tk.Toplevel()
    bt3_w.title("수정")
    bt3_w.geometry("400x200")

    def bt3_serch():
        global bt3_year_en, bt3_month_en, bt3_day_en, bt3_hour_en, bt3_item_en, bt3_money_en, combobox3, bt3_en_new
        curs.execute("select * from cal")
        rows = curs.fetchall()
        bt3_en_new=int(bt3_en.get())
        for row in rows:
            if row[0]==bt3_en_new:

                bt3_year_lb=tk.Label(bt3_w, text='년도')
                bt3_year_en=tk.Entry(bt3_w)
                bt3_year_en.insert(0, str(row[1])[:4])
                bt3_year_lb.grid(row=1, column=0)
                bt3_year_en.grid(row=1, column=1)

                bt3_month_lb=tk.Label(bt3_w, text='월')
                bt3_month_en=tk.Entry(bt3_w)
                bt3_month_en.insert(0, str(row[1])[5:7])    
                bt3_month_lb.grid(row=2, column=0)
                bt3_month_en.grid(row=2, column=1)

                bt3_day_lb=tk.Label(bt3_w, text='일')
                bt3_day_en=tk.Entry(bt3_w)
                bt3_day_en.insert(0, str(row[1])[8:10])    
                bt3_day_lb.grid(row=3, column=0)
                bt3_day_en.grid(row=3, column=1)

                bt3_hour_lb=tk.Label(bt3_w, text='시간')
                bt3_hour_en=tk.Entry(bt3_w)
                bt3_hour_en.insert(0, str(row[1])[11:13])    
                bt3_hour_lb.grid(row=4, column=0)
                bt3_hour_en.grid(row=4, column=1)

                bt3_item_lb=tk.Label(bt3_w, text='항목')
                bt3_item_en=tk.Entry(bt3_w)
                bt3_item_en.insert(0, str(row[2]))   
                bt3_item_lb.grid(row=5, column=0)
                bt3_item_en.grid(row=5, column=1)
             
                bt3_money_lb=tk.Label(bt3_w, text='금액')
                bt3_money_en=tk.Entry(bt3_w)
                bt3_money_en.insert(0, row[3]) 
                bt3_money_lb.grid(row=6, column=0)
                bt3_money_en.grid(row=6, column=1)

                combobox3 = ttk.Combobox(bt3_w, values=["수입", "지출"])
                combobox3.grid(row=7, column=0)
    def bt3_search2():                
        bt3_en_year = bt3_year_en.get()
        bt3_en_month = bt3_month_en.get()
        bt3_en_day = bt3_day_en.get()
        bt3_en_hour = bt3_hour_en.get()
        bt3_en_item = bt3_item_en.get()
        bt3_en_money = bt3_money_en.get()
        selected_item3 = combobox3.get()

        datetime_str = "{}-{}-{} {}".format(bt3_en_year, bt3_en_month, bt3_en_day, bt3_en_hour)
        
        curs.execute("update cal set datetime='{}', item='{}', money={}, sl='{}' where num={}".format(datetime_str, bt3_en_item, int(bt3_en_money), selected_item3, bt3_en_new))
        conn.commit()

    bt3_en=tk.Entry(bt3_w)
    bt3_bt=tk.Button(bt3_w, text='번호 검색', command=bt3_serch)
    bt3_en.grid(row=0, column=0)
    bt3_bt.grid(row=0, column=1)
    bt3_bt_st=tk.Button(bt3_w, text='저장', command=bt3_search2)
    bt3_bt_st.grid(row=7, column=2)



def bt4_click():
    bt4_w = tk.Toplevel()
    bt4_w.title("삭제")
    bt4_w.geometry("400x200")
    
    def delete_selected():
        selected_index = lbox.curselection()[0]  # 선택한 항목의 인덱스 가져오기
        selected_data = lbox.get(selected_index)  # 선택한 항목의 데이터 가져오기
        num_to_delete = selected_data[0]  # 선택한 항목의 번호 가져오기
        curs.execute("DELETE FROM cal WHERE num=%s", (num_to_delete,))
        conn.commit()
        lbox.delete(selected_index)  # 리스트 박스에서 선택한 항목 삭제
    
    lbox = tk.Listbox(bt4_w, width=40, height=10)
    lbox.pack()
    curs.execute("SELECT * FROM cal ORDER BY num DESC")
    rows = curs.fetchall()
    for row in rows:
        lbox.insert(0, row)

    delete_btn = tk.Button(bt4_w, text="선택 항목 삭제", command=delete_selected)
    delete_btn.pack()



def bt5_click():
    bt5_w = tk.Toplevel()
    bt5_w.title("월별 조회")
    bt5_w.geometry("400x200")

    lbox=tk.Listbox(bt5_w, width=40, height=10)
    lbox.pack()
    curs.execute("SELECT MONTH(datetime) AS Month, sl, SUM(money) AS Total FROM cal GROUP BY Month, sl ORDER BY MONTH(datetime)")
    rows = curs.fetchall()
    for row in rows:
        lbox.insert(tk.END, f"{row[0]}월, 종류: {row[1]}, 금액: {row[2]}")   


w = tk.Tk()
w.geometry("200x180")
w.title("가계부")

l_font = font.Font(family="Helvetica", size=20)

label = tk.Label(w, text='가계부', font=l_font)

label.grid(row=0, column=1)

bt1=tk.Button(w, text='입력', command=bt1_click)
bt2=tk.Button(w, text='조회', command=bt2_click)
bt3=tk.Button(w, text='수정', command=bt3_click)
bt4=tk.Button(w, text='삭제', command=bt4_click)
bt5=tk.Button(w, text='월별 조회', command=bt5_click)

bt1.grid(row=1, column=0)
bt2.grid(row=1, column=1)
bt3.grid(row=1, column=2)
bt4.grid(row=2, column=2)
bt5.grid(row=2, column=0)
t=time.localtime()


w.mainloop()
```

```python
# tkcalendar
from tkinter import *
from tkcalendar import *
import datetime

today_time = datetime.date.today()

root = Tk()
root.title("Calendar")
root.geometry("600x400")

cal = Calendar(root, selectmode="day", year=today_time.year, month=today_time.month, day=today_time.day)
cal.pack(fill="both", expand=True)

root.mainloop()
```
