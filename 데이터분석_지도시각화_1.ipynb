### [실습] : 우리나라 행정구역 경계 지도에 표시하기 : folium.GeoJson()

```python
import folium
import json

# 1. GeoJson 파일 불러오기
# file = 'skorea-provinces-geo.json' # 대한민국 지도 경계 정보
# file = 'skorea-municipalities-geo.json' # 대한민국 시군구 경계 정보
# file = 'skorea-submunicipalities-geo.json' # 대한민국 읍면동 경계 정보
# file = 'HangJeongDong_ver20230701.geojson' # 대한민국 행정동
file = 'hangjeongdong_서울특별시.geojson' # 서울특별시 행정동
data = open(file, 'r', encoding='utf-8').read()
geo_data = json.loads(data)

# 2. 지도에 GeoJson 정보를 표현하기
m = folium.Map(location=[37.559984, 126.9753071],
              zoom_start=11)
folium.GeoJson(geo_data, name='json_data').add_to(m)
m
```
