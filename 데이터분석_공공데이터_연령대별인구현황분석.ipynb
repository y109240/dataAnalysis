## 데이터 전처리

### 행정구역을 인덱스로 설정하기

```python
df.index
```

```python
# 열 이름 간단하게 자르기
df_columns = []
for i in range(len(df.columns)):
  if i == 0:
    df_columns.append('행정구역')
  else:
    df_columns.append(df.columns[i][df.columns[i].find('_') +1:])

df.columns = df_columns
df.columns
```

```python
df.head(2)
```
