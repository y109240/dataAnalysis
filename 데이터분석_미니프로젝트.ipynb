```python
# 각 구별로 동별로 동물등록수 정리
# 구별 혹은 동별로 상자그림, 지도로 보여주기
# 인프라(미용, 공원, 병원 등) 조사
import pandas as pd

```

```python
강남구 = pd.read_csv('반려동물/강남구.csv', encoding='cp949')
강남구 = 강남구[['동명', '동물 등록수']]
강남구.columns = ['동명', '동물등록수']
강남구
```

```python
강동구 = pd.read_csv('반려동물/강동구.csv', encoding='cp949')
강동구['동물등록수'] = 강동구.RFID내장형 + 강동구.RFID외장형 + 강동구.인식표
강동구 = 강동구[['읍면동', '동물등록수']]
강동구.columns = ['동명', '동물등록수']
강동구
```

```python
강북구 = pd.read_csv('반려동물/강북구.csv', encoding='cp949')
강북구['동물등록수'] = 강북구['RFID종류 내장형'] + 강북구['RFID종류 외장형'] + 강북구['RFID종류 인식표']
강북구 = 강북구[['읍면동(법정동)', '동물등록수']].iloc[:5, ]
강북구.columns = ['동명', '동물등록수']
강북구
```

```python
강서구 = pd.read_csv('반려동물/강서구.csv', encoding='cp949')
강서구['동물등록수'] = 강서구.개등록수 + 강서구.고양이등록수
강서구
```

```python
관악구 = pd.read_csv('반려동물/관악구.csv', encoding='cp949')
관악구 = 관악구[['읍면동명', '등록동물수']]
관악구.columns = ['동명', '등록동물수']
관악구
```

```python
광진구 = pd.read_csv('반려동물/광진구.csv', encoding='cp949')
광진구['등록동물수'] = 광진구['RFID종류(내장형)'] + 광진구['RFID종류(외장형)'] + 광진구['RFID종류(인식표)']
광진구 = 광진구[['읍면동(법정동)', '등록동물수']]
광진구.columns = ['동명', '등록동물수']
광진구
```

```python
금천구 = pd.read_csv('반려동물/금천구.csv', encoding='cp949')
금천구 = 금천구[['법정동', '개체수(전체)']]
금천구.columns = ['동명', '등록동물수']
금천구
```

```python
노원구 = pd.read_csv('반려동물/노원구.csv', encoding='cp949')
노원구['등록동물수'] = 노원구['RFID종류(내장형)'] + 노원구['RFID종류(외장형)'] + 노원구['RFID종류(인식표)']
노원구 = 노원구[['읍면동(법정동)', '등록동물수']].iloc[-6:-1, ]
노원구.columns = ['동명', '등록동물수']
노원구
```
