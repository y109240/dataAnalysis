```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
강남구 = pd.read_csv('미니프로젝트_데이터/강남구.csv', encoding='cp949')
강남구 = 강남구[['동명', '동물 등록수']]
강남구.columns = ['동명', '등록동물수']
강남구
```

```python
강동구 = pd.read_csv('미니프로젝트_데이터/강동구.csv', encoding='cp949')
강동구['등록동물수'] = 강동구.RFID내장형 + 강동구.RFID외장형 + 강동구.인식표
강동구 = 강동구[['읍면동', '등록동물수']]
강동구.columns = ['동명', '등록동물수']
강동구
```

```python
강북구 = pd.read_csv('미니프로젝트_데이터/강북구.csv', encoding='cp949')
강북구['등록동물수'] = 강북구['RFID종류 내장형'] + 강북구['RFID종류 외장형'] + 강북구['RFID종류 인식표']
강북구 = 강북구[['읍면동(법정동)', '등록동물수']].iloc[:4, ]
강북구.columns = ['동명', '등록동물수']
강북구
```

```python
강서구 = pd.read_csv('미니프로젝트_데이터/강서구.csv', encoding='cp949')
강서구['고양이등록수'].fillna(0, inplace=True) # 결측치 변환
강서구['등록동물수'] = 강서구.개등록수 + 강서구.고양이등록수
강서구 = 강서구[['법정동', '등록동물수']]
강서구.columns = ['동명', '등록동물수']
강서구
```

```python
관악구 = pd.read_csv('미니프로젝트_데이터/관악구.csv', encoding='cp949')
관악구 = 관악구[['읍면동명', '등록동물수']]
관악구.columns = ['동명', '등록동물수']
관악구
```

```python
광진구 = pd.read_csv('미니프로젝트_데이터/광진구.csv', encoding='cp949')
광진구['등록동물수'] = 광진구['RFID종류(내장형)'] + 광진구['RFID종류(외장형)'] + 광진구['RFID종류(인식표)']
광진구 = 광진구[['읍면동(법정동)', '등록동물수']]
광진구.columns = ['동명', '등록동물수']
광진구
```

```python
구로구 = pd.read_csv('미니프로젝트_데이터/구로구.csv', encoding='cp949')
구로구['등록동물수'] = 구로구['내장형(RFID)'] + 구로구['외장형(RFID)'] + 구로구['인식표']
구로구 = 구로구[['법정동', '등록동물수']]
구로구.columns = ['동명', '등록동물수']
구로구
```

```python
금천구 = pd.read_csv('미니프로젝트_데이터/금천구.csv', encoding='cp949')
금천구 = 금천구[['법정동', '개체수(전체)']]
금천구.columns = ['동명', '등록동물수']
금천구
```

```python
노원구 = pd.read_csv('미니프로젝트_데이터/노원구.csv', encoding='cp949')
노원구 = 노원구[['읍면동(법정동)', '합계']].iloc[-6:-1, ]
노원구.columns = ['동명', '등록동물수']
노원구
```

```python
도봉구 = pd.read_csv('미니프로젝트_데이터/도봉구.csv', encoding='cp949')
도봉구.columns = ['동명', '등록동물수']
도봉구['등록동물수'] = 도봉구.등록동물수.str.replace(",","").str.replace("건", "").astype('int64')
도봉구
```

```python
동대문구 = pd.read_csv('미니프로젝트_데이터/동대문구.csv', encoding='cp949')
동대문구['등록동물수'] = 동대문구.등록형태_내장형RFID + 동대문구.등록형태_외장형RFID + 동대문구.등록형태_인식표RFID
동대문구 = 동대문구[['읍면동(법정동)', '등록동물수']]
동대문구.columns = ['동명', '등록동물수']
동대문구
```

```python
동작구 = pd.read_csv('미니프로젝트_데이터/동작구.csv', encoding='cp949')
동작구 = 동작구[['법정동', '반려동물총계']]
동작구.columns = ['동명', '등록동물수']
동작구
```

```python
성동구 = pd.read_csv('미니프로젝트_데이터/성동구.csv', encoding='cp949')
성동구 = 성동구[['법정동', '등록수']]
성동구.columns = ['동명', '등록동물수']
성동구
```

```python
중구 = pd.read_csv('미니프로젝트_데이터/중구.csv', encoding='cp949')
중구['등록동물수'] = round(중구.동물소유자수 * 중구.동물소유자당동물등록수)
중구 = 중구[['읍면동(법정동)', '등록동물수']]
중구.columns = ['동명', '등록동물수']
중구
```

```python
중랑구 = pd.read_csv('미니프로젝트_데이터/중랑구.csv', encoding='cp949')
중랑구 = 중랑구[['읍면동(법정동)', '합계']]
중랑구.columns = ['동명', '등록동물수']
중랑구
```

```python
# 구별로 총계, 데이터프레임, 지도표현
total_df = pd.DataFrame(columns=['구', '등록동물수'])
구 = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구',
    '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구',
    '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

등록동물수 = [
  강남구.등록동물수.sum(), 강동구.등록동물수.sum(), 강북구.등록동물수.sum(), 강서구.등록동물수.sum(),
  관악구.등록동물수.sum(), 광진구.등록동물수.sum(), 구로구.등록동물수.sum(), 금천구.등록동물수.sum(),
  노원구.등록동물수.sum(), 도봉구.등록동물수.sum(), 동대문구.등록동물수.sum(), 동작구.등록동물수.sum(),
  마포구.등록동물수.sum(), 16944, 0, 성동구.등록동물수.sum(), 성북구.등록동물수.sum(),
  송파구.등록동물수.sum(), 0, 영등포구.등록동물수.sum(), 용산구.등록동물수.sum(), 0,
  종로구.등록동물수.sum(), 중구.등록동물수.sum(), 중랑구.등록동물수.sum()]

total_df['구'] = 구
total_df['등록동물수'] = 등록동물수
total_df
```

```python
# 막대 그래프
total_df_sort = total_df.sort_values('등록동물수', ascending=False)
plt.figure(figsize=(15, 5))
plt.bar(total_df_sort['구'], total_df_sort['등록동물수'])
plt.title('서울시 구별 등록동물수', fontsize=20)
```

```python
# 지도 시각화
import folium
import json

file = 'seoul_geo_simple.json'
data = open(file, 'r', encoding='utf-8').read()
geo_data = json.loads(data) 

m = folium.Map(location=[37.566345, 126.977893], zoom_start=11, tiles="OpenStreetMap")

folium.Choropleth(
    geo_data=geo_data,
    name='인구수',
    data=total_df,
    columns=('구', '등록동물수'),
    key_on='feature.properties.name',
    fill_color='BuPu',  # viridis
    legend_name='total',
    line_weight=2,
    highlight=True
).add_to(m)
m
```

```python
# 서울시 구별 인구분포
file = '미니프로젝트_데이터/202403주민등록인구.csv'
df = pd.read_csv(file, encoding = 'cp949')

# 데이터 전처리
df_t = df[ ['행정구역','2024년03월_총인구수','2024년03월_남자 인구수','2024년03월_여자 인구수']]
df_t.columns = ['행정구역','total', '남자인구', '여자인구']
df_t.drop(0, axis=0, inplace=True)
df_t.insert(2, '시군구', df_t['행정구역'].str.split(' ').str[2])
df_t['total'] = df_t['total'].str.replace(',','').apply('int64')
df_t['남자인구'] = df_t['남자인구'].str.replace(',','').apply('int64')
df_t['여자인구'] = df_t['여자인구'].str.replace(',','').apply('int64')
df_t = df_t[['시군구', 'total']].sort_values('시군구')
df_t
```

```python
df_t_sort = df_t.sort_values('total', ascending=False)
plt.figure(figsize=(15, 5))
plt.bar(df_t_sort['시군구'], df_t_sort['total'])
plt.title('서울시 구역별 인구분포', fontsize=20)
```
