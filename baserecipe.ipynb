```python
# 기본 레시피
from django.db import models

class BaseRecipe( models.Model ):
    brecipe_id = models.AutoField( verbose_name="기본레시피아이디", primary_key=True )    
    brecipe_name = models.CharField( max_length=50, verbose_name="요리명",  null=False )
    category = models.CharField( max_length=50, verbose_name="카테고리", null=False )
    img_url = models.CharField( max_length=100, verbose_name="이미지url", null=False)
    level = models.CharField( max_length=10, verbose_name="난이도", null=False )
    ingredient = models.TextField( verbose_name="재료", null=False )
    cookstep = models.TextField(verbose_name="요리순서", null=False )
    likes = models.IntegerField( verbose_name="좋아요", default=0 )
    hits = models.IntegerField( verbose_name="조회수", default=0 )
  
```

```python
# csv파일을 json으로 변환
# json파일을 loader로 만들기
import json

with open("brecipe_data.json", "r", encoding="utf-8") as f:
  brecipes = json.load(f)

new_list = []
for brecipe in brecipes:
  new_data = {"model" : "recipe.baserecipe"}
  # pk는 autofield
  new_data["fields"] = brecipe
  new_list.append(new_data)

with open("baserecipe.json", "w", encoding="utf-8") as f:
  json.dump(new_list, f, ensure_ascii=False, indent=2)
```
